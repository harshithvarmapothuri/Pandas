# Selecting Particular DataType Data from a DF

## Select Objects

```python
df.dtypes=='object'
```

```python
df.dtypes[df.dtypes=='object']
```

```python
df.dtypes[df.dtypes=='object'].index
```

```python
obs
```

## Selecting Numericals

```python
df.dtypes[(df.dtypes=='int64') | (df.dtypes=='float64')]
```

```python
df.dtypes[(df.dtypes=='int64') | (df.dtypes=='float64')].index
```

```python
nums=df[df.dtypes[(df.dtypes=='int64') | (df.dtypes=='float64')].index]
```

# Filtering a DF

## Single Filter

```python
df[df["Survived"]==1]
```

```python
df[(df["Age"]>65) & (df["Fare"]>70)]
```
